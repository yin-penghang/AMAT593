# Matrix Plots

Use Matrix Plots to plot data as color-encoded matrices and to indicate clusters within the data.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Heatmap

For data already in a matrix form, the sns.heatmap function basically just colors it. For example:

```python
tips.head()
```

```python
len(tips)
```

```python
sns.heatmap(tips[['total_bill']])
```

```python
# Matrix form for correlation data
tips[['total_bill','tip','size']].corr()
```

```python
sns.heatmap(tips[['total_bill','tip','size']].corr())
```

```python
ax = sns.heatmap(tips[['total_bill','tip','size']].corr(),cmap='coolwarm',annot=True) #anontation
ax.set_aspect("equal")
```
