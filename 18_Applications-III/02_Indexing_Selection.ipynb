In this lecture we will discuss how to select elements or groups of elements from an array.

## Bracket Indexing and Selection
The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
#Get a value at an index
arr[8]
```

```python
#Get values in a range
arr[0:5]
```

```python
#Get values in a range
arr[1:] # same as arr[1:end]
```

```python
arr[:5] # same as arr[0:5]
```

```python
arr[0:5]
```

## Broadcasting

Numpy arrays differ from a normal Python list because of their ability to broadcast:

```python
arr
```

```python
#Setting a value with index range (Broadcasting)
arr[0:5]=10

#Show
arr
```

```python
# Reset array, we'll see why I had to reset in  a moment
arr = np.arange(0,11)

#Show
arr
```

```python
#Important notes on Slices
slice_of_arr = arr[0:5]

#Show slice
slice_of_arr
```

```python
#Change Slice
slice_of_arr[:]=99

#Show Slice again
slice_of_arr
```

Now note the changes also occur in our original array.

```python
arr
```

Data is not copied, it's a view of the original array. This avoids memory problems.

```python
arr = np.arange(0,11)
arr
```

```python
#To get a copy, need to be explicit

arr_copy = arr.copy()

arr_copy
```

```python
arr_copy[0:5] = 99
arr_copy
```

```python
# the original array is unchanged
arr
```

## Indexing a 2D array (matrices)

The general format is **arr_2d[row][col]** or **arr_2d[row,col]**. I recommend usually using the comma notation for clarity.

```python
arr_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))

#Show
arr_2d
```

```python
#Indexing row
arr_2d[0]
```

```python
arr_2d[0:2]
```

```python
# Format is arr_2d[row][col] or arr_2d[row,col]

# Getting individual element value
arr_2d[1][0]
```

```python
# Getting individual element value
arr_2d[1,0]
```

```python
arr_2d
```

```python
# 2D array slicing

#Shape (2,2) from top left corner
arr_2d[:2,:2]
```

```python
# top right corner
arr_2d[:2,1:]
```

```python
#Shape bottom row
arr_2d[2]
```

```python
#Shape bottom row
arr_2d[2,:]
```

```python
arr_2d[:,0] # first column
```

```python
arr_2d[:,0].shape
```

```python
arr_2d
```

```python
arr_2d[:,:2] # first 2 columns
```

```python
arr_2d[1:,:] # last 2 rows
```

```python
arr_2d
```

```python
arr_2d[:,0].shape # 1-d array
```

```python
arr_2d[:,0]
```

```python
arr_2d[:,0].reshape(3,1) #  convert it into 2-d array of same dimension 
```

```python
arr_2d[:,0].reshape(3,1).shape 
```

### Fancy Indexing

Fancy indexing allows you to select entire rows or columns out of order, to show this, let's quickly build out a numpy array:

```python
#Set up matrix
arr2d = np.zeros((10,5))

arr2d
```

```python
arr2d.shape
```

```python
# # rows of array
num_row = arr2d.shape[0]
print(num_row)
```

```python
num_col = arr2d.shape[1]
print(num_col)
```

```python
#Set up array

for i in range(num_row):
    arr2d[i] = i 
arr2d
```

Fancy indexing allows the following

```python
arr2d[[2,4,6,8]]
```

```python
#Allows in any order
arr2d[[6,4,2,7]]
```

## Selection

Let's briefly go over how to use brackets for selection based off of comparison operators.

```python
arr = np.arange(1,11)
arr
```

```python
arr > 4
```

```python
bool_arr = arr>4
```

```python
bool_arr
```

```python
arr[arr>4]
```

```python
arr[arr>2]
```

```python
x = 2
arr[arr>x]
```

```python
arr[arr<=4]
```

```python
arr[arr==4]
```

```python
arr[(arr>3) & (arr<8)]
```
